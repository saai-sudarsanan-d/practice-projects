```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from rdkit import Chem
from rdkit.Chem import AllChem, Descriptors, Draw
%matplotlib inline
```

```
smi = pd.read_csv('Smiles.csv')
smi
```

```
def show_mol(mol_smile):
  Draw.MolToMPL(Chem.MolFromSmiles(mol_smile),size = (200,200))
  plt.title('{}'.format(smi['Drug_Name'][i]))
  plt.show()
  plt.clf()
for i in range(30):
  mol_smile = smi['Smile'][i]
  show_mol(mol_smile)
```
